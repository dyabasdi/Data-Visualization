**Q. 7b (10 points)**.  
Create boxplots showing the immigration of all countries over the years 1980 - 2013.
* Your graph should contain 134 boxplots, one boxplot per country, and each boxplot will display the immigration data from 1980 - 2013 for that country.
* You may need to use `pd.melt()` or `pd.transpose()` to arrange your data in the correct format for graphing.
* For graphing, we recommend using seaborn, but matplotlib is also acceptable. You should also create appropriate labels and modify them (if needed) to make them readable.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_melted = df.reset_index().melt(
    id_vars=['Country'], 
    value_vars=df.loc[:, '1980':'2013'].columns.tolist(),  # Convert Index to list    var_name='Year', 
    value_name='Immigration'
)

# Create boxplots
plt.figure(figsize=(30, 20))
sns.boxplot(x='Country', y='Immigration', data=df_melted)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Immigration')
plt.title('Immigration from 1980 to 2013 by Country')
plt.show()
```

**Q. 8 (20 points)**    
For the countries in the Region named as Caribbean, find the total immigration in the two decades (1980 to 1989) and (1990 - 1999). Print the decades as (Eighties and Nineties) and the total immigration for these decades from each of the Caribbean countries.

```python
# Filter for Caribbean countries
caribbean_mask = df['Region'] == 'Caribbean'

# Calculate total immigration for 1980-1989 and 1990-1999 using .loc on the original DataFrame
df.loc[caribbean_mask, 'Eighties'] = df.loc[caribbean_mask, '1980':'1989'].sum(axis=1)
df.loc[caribbean_mask, 'Nineties'] = df.loc[caribbean_mask, '1990':'1999'].sum(axis=1)

# Extract the filtered DataFrame for printing
caribbean_countries = df[caribbean_mask]

# Print the results
for country, row in caribbean_countries.iterrows():
    print(f"{country}: Eighties - {row['Eighties']}, Nineties - {row['Nineties']}")
```

**Q. 9 (20 points)**.  
The Scandinavian countries are Denmark, Norway, and Sweden. For each of the years (from 1980 to 2013), find the difference between the country with the highest immigration and the country with the lowest immigration. Print the year and the difference.

```python
# Filter for Scandinavian countries
scandinavian_countries = df.loc[['Denmark', 'Norway', 'Sweden']]

# Calculate the difference for each year
for year in df.columns[3:-1]:
    max_immigration = scandinavian_countries[year].max()
    min_immigration = scandinavian_countries[year].min()
    difference = max_immigration - min_immigration
    print(f"{year}: {difference}")
```

**Q. 10 (20 points)**.  
Find the countries in Africa that had the highest and lowest immigration for each of the years. Print the year, name of the country with highest immigration, immigration number, name of the country with the lowest immigration, and the immigration number. 

```python
# Filter for African countries
african_countries = df[df['Continent'] == 'Africa']

# Find the highest and lowest immigration for each year
for year in african_countries.loc[:, '1980':'2013'].columns:
    max_country = african_countries[year].idxmax()
    min_country = african_countries[year].idxmin()
    max_immigration = african_countries.loc[max_country, year]
    min_immigration = african_countries.loc[min_country, year]
    print(f"{year}: {max_country} ({max_immigration}), {min_country} ({min_immigration})")
```
