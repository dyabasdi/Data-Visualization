#### P-value

The probability that the correlation between two variables in statistically significant.
* p < 0.001:    strong evidence that the correlation coefficient is significant.
* p < 0.05:     moderate evidence that the correlation coefficient is significant.
* p < 0.1:      weak evidence that the correlation coefficient is significant.
* p > 0.1:      no evidence that the correlation coefficient is significant.

```python
# import scipy stats library
from scipy import stats
```

```python
# compute the correlation coefficient and the p-value for engine size
p_coef, p_val = stats.pearsonr(df['engine-size'], df['price'])
print(f"P_Coef: {p_coef}\n P_Val: {p_val}")
```

Tight correlation - makes sense

#### Analysis of Variance (ANOVA)
Analysis of Variance is a statistical method to test whether there are significant differences between two or more groups. ANOVA returns two parameters:
* F-test Score
* P-Value

If our price variable is strongly correlated with the variable that we are analyzing, then we will expect ANOVA to return a large F-test score and a small p-value.


```python
# check if there is a group 2
print(group_2)
```

```python
# make another group from group 2
group_3 = group_2[['drive-wheels', 'price']].groupby(['drive-wheels'])
group_3.head(3)
```

```python
group_3.get_group('4wd')['price']
```

```python
# run ANOVA on the different groups
f_val, p_val = stats.f_oneway(group_3.get_group('fwd')['price'],
                              group_3.get_group('4wd')['price'],
                              group_3.get_group('rwd')['price'],
                              )
print(f"ANOVA Results:\n - F_val: {f_val}\n - P_val: {p_val}")
```

#### Linear Regression Model

y = a + bx

```python
# import scikit learn
from sklearn.linear_model import LinearRegression
```

```python
# create a linear regression object
lm = LinearRegression()

# use engine-size first

X = df[['engine-size']]
Y = df['price']
lm.fit(X, Y)
```

```python
# get slope and intercept
print(f"Slope: {lm.coef_}\nIntercept: {lm.intercept_}")
lm.score(X, Y)
```

Equation:

y = 166.86x - 7963.34

Price = 166.86 * engine-size - 7963.34

```python
sns.regplot(x = 'engine-size', y= 'price', data = df)
```

```python
# compare with actual values by creating residual plot
sns.residplot(x = df['engine-size'], y = df['price'])
```

#### Coefficient of determination

R^2 Value indicates how well a model fits the data and predicts outcomes.


```python
# coefficient of determination
print(f"R^2: {lm.score(X, Y)}")
```

### Polynomial Fit

```python
# function to plot data
def plotPoly(model, x, y, Name):
    x_new = np.linspace(1400, 5000, 1000)
    y_new = model(x_new)

    plt.plot(x, y, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')
    plt.show()
```

```python
# fit a quadratic to the data
x = df['curb-weight']
y = df['price']
f = np.polyfit(x, y, 2)
p = np.poly1d(f)
plotPoly(p, x, y, 'Curb Weight')
```
