**Q. 2 (20 points)**  
For each year, find the maximum number of immigrants and the country that had the maximum number of immigrants. Create a new dataframe having three columns - year, country, and number of immigrants. For a given year, if there are two or more countries having the same maximum number of immigrants, then choose that country that comes first in alphabetical order.

```python
# remove total immigration
temp_df = df
temp_df = temp_df[temp_df['Country'] != 'total_immigration']

# init dataframe
df_immigrants = pd.DataFrame(columns=['Year', 'Country', 'Immigrants'])

for year in map(str, np.arange(1980, 2014)):
    # get max number of immigrants for that year and the countries that have this value
    max_value = temp_df[year].max()
    max_countries = temp_df.loc[temp_df[year] == max_value, 'Country']

    # don't add if no max
    if not max_countries.empty:
        # sort array to get it in alphabetical order
        max_country = sorted(max_countries)
        df_immigrants.loc[len(df_immigrants)] = [int(year), max_country[0], max_value]

print(df_immigrants)
```
